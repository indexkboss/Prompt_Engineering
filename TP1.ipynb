````python
#la tokenisation c'est la premiere étape de la NLP 
#c'est le processus de diviser un texte en unités plus petites appelées tokens.
#ces tokens peuvent être des mots, phrases, des caractères.., en fonction de la méthode de tokenisation
#utilisée. 

#c'est une préparation du texte pour les étapes suivantes telles que: 
#l'analyse syntaxique, la reconnaissance d'entités nommées, et la modélisation du langage.

#Il existe plusieurs méthodes de tokenisation
````

````python
#on veut savoir le nb de tokens dans un prompt pour ??(facturation) on utilise librairie tiktoken:
import tiktoken
````

````python
#methode 1: specifier le model pour lequel on veut connaitre le tokennizer
tokennizer = tiktoken.encoding_for_model("gpt-4o")
print(tokennizer.name)#affiche le nom du tokennizer utilisé pour ce model
````

````python
#methode 2 : on done le nom directe de loutil de tokenisation du modele
tokennizer = tiktoken.get_encoding("o200k_base")
````

````python
prompt = "au clair de la lune, mon ami pierrot, prête-moi ta plume pour écrire un mot"
````

````python
tokens = tokennizer.encode(prompt)
````

````python
print(tokens) #affiche la liste des tokens, chaque token est représenté par un entier.
````

````python
print(len(tokens))
````

````python
print("1:\n")
for token in tokens:
    print(tokennizer.decode_single_token_bytes(token))#format bytes (brut)
    # Prend UN SEUL token → retourne des bytes (brut), pas une chaîne
print("2:\n")
for token in tokens:
    print((tokennizer.decode_single_token_bytes(token).decode("utf-8")),end="|") #format texte, end pour separer par |

print("3:\n")
print(tokennizer.decode(tokens))
# Prend une liste de tokens → retourne une chaîne de caractères (str)
````

````python
#on a besoin de connaitre le nb de tokens 
#cette fonction :
def tokens_count(prompt:str, model:str="o200k_base"):#o200k_base est le modele de tokenisation par defaut 
    tokennizer = tiktoken.get_encoding(model)#on récupère le "tokenizer" du modèle choisi
    tokens = tokennizer.encode(prompt)#le decoupement du prompt en tokens
    return len(tokens)
````

````python
print(tokens_count(prompt))
````

````python
from huggingface_hub import InferenceClient # Hugging Face API client (this is what actually calls the model)
from PIL import Image # to manipulate images  
from io import BytesIO

from IPython.display import Image
import base64 
````

````python
# since we dont have access to gpt-5.2, we can use a workaround with HuggingFace Inference API to generate images with a free model
from huggingface_hub import InferenceClient # Hugging Face API client (this is what actually calls the model)
from PIL import Image # to manipulate images
from io import BytesIO


HF_TOKEN = os.getenv("HF_TOKEN")

if not HF_TOKEN:
    raise ValueError("Missing HF_TOKEN")

# creating a connection to Hugging Face’s inference API.
client = InferenceClient(
    provider="hf-inference",
    api_key=HF_TOKEN
)

def generate_image(prompt):# takes an input prompt, calls the model, 
    image = client.text_to_image(
        prompt,
        model="black-forest-labs/FLUX.1-schnell"
    )

    # Convert PIL image → base64 (to mimic OpenAI response(Turn a PIL image object into a string that can be sent in JSON.))
    buffer = BytesIO()
    image.save(buffer, format="PNG")
    base64_image = base64.b64encode(buffer.getvalue()).decode("utf-8")

    # mimic OpenAI tool response structure
    return {
        "content_blocks": [
            {
                "type": "image",
                "base64": base64_image
            }
        ]
    }
# -------------------
# TEST
# -------------------
resp = generate_image("a cute monkey dancing wearing a skirt and a hat, in the style of van gogh")

img = Image.open(BytesIO(base64.b64decode(resp["content_blocks"][0]["base64"])))
display(img)
````

## use case: Analyse de sentiment des avis sur des ordinateurs portables

````python
system_message = """
Vous êtes un expert dans l'analyse des sentiments.
Effectuez une analyse de sentiments basée sur les aspects des avis concernant les ordinateurs portables présentés en entrée.

Chaque avis peut comporter un ou plusieurs des aspects suivants : screen, keyboard, pad.

Pour chaque avis présenté en entrée :
- Identifiez la présence d'au moins un des trois aspects (screen, keyboard, pad).
- Attribuez une polarité de sentiment (positive, negative ou neutral) à chaque aspect.
- Organisez votre réponse dans un objet JSON avec les en-têtes suivants :
    - category: [liste des aspects]
    - polarity: [liste des polarités correspondantes pour chaque aspect]

Si l'un des aspects n'est pas présent dans l'avis de l'utilisateur, tu supposes que la polarité est neutre.
"""
````

````python
llm6 = ChatOpenAI(
    model="openai/gpt-4o-mini",
    base_url="https://openrouter.ai/api/v1",
        max_tokens=500
    )
````

````python
resp = llm6.invoke(input=[
    SystemMessage(content=system_message),
    HumanMessage(content="j ai beaucoup aimé l ecran, la souris n est pas bonne, et le clavier ma fih ma ytchaf ")
])
````

````python
print((resp.content))
````
