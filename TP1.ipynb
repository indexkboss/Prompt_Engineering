```python
#la tokenisation c'est la premiere étape de la NLP 
#c'est le processus de diviser un texte en unités plus petites appelées tokens.
#ces tokens peuvent être des mots, phrases, des caractères.., en fonction de la méthode de tokenisation
#utilisée. 

#c'est une préparation du texte pour les étapes suivantes telles que: 
#l'analyse syntaxique, la reconnaissance d'entités nommées, et la modélisation du langage.

#Il existe plusieurs méthodes de tokenisation
```

```python
#on veut savoir le nb de tokens dans un prompt pour ??(facturation) on utilise librairie tiktoken:
import tiktoken
```

```python
#methode 1: specifier le model pour lequel on veut connaitre le tokennizer
tokennizer = tiktoken.encoding_for_model("gpt-4o")
print(tokennizer.name)#affiche le nom du tokennizer utilisé pour ce model
```

```python
#methode 2 : on done le nom directe de loutil de tokenisation du modele
tokennizer = tiktoken.get_encoding("o200k_base")
```

```python
prompt = "au clair de la lune, mon ami pierrot, prête-moi ta plume pour écrire un mot"
```

```python
tokens = tokennizer.encode(prompt)
```

```python
print(tokens) #affiche la liste des tokens, chaque token est représenté par un entier.
```

```python
print(len(tokens))
```

```python
print("1:\n")
for token in tokens:
    print(tokennizer.decode_single_token_bytes(token))#format bytes (brut)
    # Prend UN SEUL token → retourne des bytes (brut), pas une chaîne
print("2:\n")
for token in tokens:
    print((tokennizer.decode_single_token_bytes(token).decode("utf-8")),end="|") #format texte, end pour separer par |

print("3:\n")
print(tokennizer.decode(tokens))
# Prend une liste de tokens → retourne une chaîne de caractères (str)
```

```python
#on a besoin de connaitre le nb de tokens 
#cette fonction :
def tokens_count(prompt:str, model:str="o200k_base"):#o200k_base est le modele de tokenisation par defaut 
    tokennizer = tiktoken.get_encoding(model)#on récupère le "tokenizer" du modèle choisi
    tokens = tokennizer.encode(prompt)#le decoupement du prompt en tokens
    return len(tokens)
```

```python
print(tokens_count(prompt))
```

## Interagir avec un LLM OpenAI (need api key):

```python
#uv add openai, langchain, langchain-openai, ollama, langchain-ollama, langchain-groq.
from langchain_openai import ChatOpenAI

#import os #creer une variable d'environnement pour stocker la API key de manière not sécurisée
# os.environ['OPENAI_API_KEY'] = "..."

from dotenv.ipython import load_dotenv
load_dotenv(override=True) #pour charger les variables d'environnement à partir du fichier .env
```

```python
import os
print(os.getenv("var_env_name")) #pour vérifier que la variable d'environnement est bien chargée
```

```python
llm1 = ChatOpenAI(
    model="openai/gpt-oss-20b:free",
    temperature=0.9,
# proche de 1 : plus creatif, a chaque fois reponse diff, 
# proche de 0 : plus factuel, a chaque fois reponse similaire
    base_url="https://openrouter.ai/api/v1"#car we're using openRouter
)
```

```python
resp = llm1.invoke(input=[
    {"role":"system","content":"you are a helpful assistant"},
    {"role":"user","content":"what's an ai agent?"}
])
#input = sys message, fewShot examples, user input.
```

```python
from IPython.display import Markdown #pour afficher la reponse de manière plus lisible
```

```python
print(display(Markdown(resp.content)))
```
